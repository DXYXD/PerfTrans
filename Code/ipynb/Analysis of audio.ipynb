```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['figure.figsize']=(8.0,6.0)
pd.set_option("display.max_rows", 6000)
```

```python
csv_path = 'D:\\Academic_work\\00PerfTransfer\\File\\csv\\'
csv_file = pd.read_csv(csv_path + 'anal_new.csv')
leng = len(csv_file)
```

```python
colnames = list(csv_file.columns)
colnames
```

### DB

```python
db_diff = csv_file.sort_values(by=['pitch'])['db(lab) - db(audi)'].tolist()
db_pm = [float('%.2f' % np.mean(db_diff[i:i+68])) for i in range(0,leng,68)]
```

```python
x = [str(i) for i in list(range(21,109))]
```

```python
import pyecharts.options as opts
from pyecharts.charts import Scatter
from pyecharts.globals import ThemeType

scatter = (
        Scatter(init_opts=opts.InitOpts(theme=ThemeType.WESTEROS))
        .add_xaxis(x)
        .add_yaxis('average diff', db_pm)
        .set_global_opts(
            title_opts=opts.TitleOpts(title='Average DB diff with each pitch'),
            xaxis_opts=opts.AxisOpts(name='Pitch'),
            yaxis_opts=opts.AxisOpts(name='Diff'),
            visualmap_opts=opts.VisualMapOpts(min_=-6, max_=12),
            )
        .set_series_opts(
            label_opts=opts.LabelOpts(is_show=True),
            markline_opts=opts.MarkLineOpts(
                data=[opts.MarkLineItem(y=1.47, name="diff average of pitch 21(A1)-47(B3)"),
                     opts.MarkLineItem(y=4.43, name="diff average of pitch 48(C4)-74(D6)"),
                     opts.MarkLineItem(y=-1.52, name="diff average of pitch 75(D#6)-108(C9)")]
            ),
        )
)

scatter.render_notebook()
```

```python
db_diff2 = csv_file.sort_values(by=['dur','pitch','vel'])['db(lab) - db(audi)'].tolist()
scatter = (
        Scatter(init_opts=opts.InitOpts(theme=ThemeType.WESTEROS))
        .add_xaxis(list(range(88*17*4+50)))
        .add_yaxis('dur diff', db_diff2, symbol_size = 5)
        .set_global_opts(
            title_opts=opts.TitleOpts(title='DB diff with each pitch'),
            xaxis_opts=opts.AxisOpts(name='Sounds'),
            yaxis_opts=opts.AxisOpts(name='Diff'),
            visualmap_opts=opts.VisualMapOpts(min_=-25, max_=25),
            )
        .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False)
        )
)

scatter.render_notebook()
```

### Duration

```python
dur_diff = csv_file.sort_values(by=['dur','pitch','vel'])['dur(lab) - dur(audi)'].tolist()
```

```python
scatter = (
        Scatter(init_opts=opts.InitOpts(theme=ThemeType.WESTEROS))
        .add_xaxis(list(range(88*17*4+50)))
        .add_yaxis('dur diff', dur_diff, symbol_size = 5)
        .set_global_opts(
            title_opts=opts.TitleOpts(title='duration diff with each pitch'),
            xaxis_opts=opts.AxisOpts(name='Sounds'),
            yaxis_opts=opts.AxisOpts(name='Diff'),
            visualmap_opts=opts.VisualMapOpts(min_=-1.5, max_=1.5),
            )
        .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False)
        )
)

scatter.render_notebook()
```

```python
v = [dur_diff[:1496], dur_diff[1496:1496*2], dur_diff[1496*2:1496*3], dur_diff[1496*3:]]
for i in range(4):
    v[i] = [v[i][:614], v[i][614:1107], v[i][1107:]]
v1 = [[v[0][i], v[1][i], v[2][i], v[3][i]] for i in range(3)]
```

```python
from pyecharts import options as opts
from pyecharts.charts import Boxplot

boxplot = (
        Boxplot(init_opts=opts.InitOpts(theme=ThemeType.WESTEROS))
        .add_xaxis(["0.1", "0.3", "0.7", "1"])
        .add_yaxis("pitch 21(A1)-56(G#3)", Boxplot.prepare_data(v1[0]))
        .add_yaxis("pitch 57(A4)-85(C#7)", Boxplot.prepare_data(v1[1]))
        .add_yaxis("pitch 86(D7)-108(C9)", Boxplot.prepare_data(v1[2]))
        .set_global_opts(title_opts=opts.TitleOpts(title="Duration"),
                        xaxis_opts=opts.AxisOpts(name='Duration'),
                        yaxis_opts=opts.AxisOpts(name='Diff'),)
)

boxplot.render_notebook()
```
