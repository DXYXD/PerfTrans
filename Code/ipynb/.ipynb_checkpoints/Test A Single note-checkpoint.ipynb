```python
import IPython.display
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
def compute_delay(frame):
    
    '''
    second
    '''
    
    sr = 48000
    threshold = 0.003
    
    length = sr//100
    j = 10
    tp = threshold+1
    while(tp > threshold):
        j+=1
        tp = np.sqrt(np.var((frame[j*length:j*length+length])))
    delay = float(j)/100
        
    return delay

def compute_db(frame):
    ref = 0.00002
    t = 20
    
    energy = np.array(frame ** 2)
    onset = np.where(energy == max(energy))[0][0] # the index of the maximum energy
    tp = np.sqrt(np.mean(energy[onset:(onset + t)])) # # root mean square amplitude of the Ô¨Årst 10ms after the peak
    decibel=librosa.amplitude_to_db([tp], ref)[0]
        
    return tp, decibel
```

```python
def loads(file_path, notetype, pitch, vel_level,vel, dur_level,dur):
    ''' sr = 48000 '''
    sr = 48000
    name = 'dictwave_'+notetype+'_pitch-'+str(pitch)+'_vel-'+str(vel[vel_level])+'_dur-'+str(dur[dur_level])+'_idx-'+str(dur_level+4*vel_level+4*len(vel)*(pitch-21))
    load_dict = np.load(file_path+'\\'+name+'.npy')
    print('type:',notetype, '  ', 'pitch:', pitch, '  ', 'velocity:', vel[vel_level], '  ', 
          'duration:', dur[dur_level], '  ', 'index:', dur_level+4*vel_level+4*len(vel)*(pitch-21))
    info = {'type': notetype, 'sr': sr, 'pitch':pitch, 'velocity':vel[vel_level], 
            'duration':dur[dur_level], 'index':dur_level+4*vel_level+4*len(vel)*(pitch-21)}
    data = load_dict
    
    return data, info
```

```python
def wave(recordtype, notetype, pitch, vel_n, dur_n):
    '''
    recordtype = new-1/old-0
    notetype = audi-1/lab-0
    '''
    if recordtype == 1:
        file_path = 'G:\\00_Academic\\MusicXLab\\npy\\2019summer'
        vel = list(range(1,127,8))+[127]
        dur = [0.1, 0.3, 0.7, 1]
    elif recordtype == 0:
        file_path = 'G:\\00_Academic\\MusicXLab\\npy\\maoran'
        dur = [0.02, 0.15, 0.3, 0.7]
        vel = list(range(7,128,8))

    vel_level = int(round(vel_n // (vel[1]-vel[0])))
    dur_level = int(round(dur_n // (max(dur)/len(dur))))
    wave, info = loads(file_path, notetype, pitch, vel_level, vel, dur_level, dur)
    
    return wave, info
```

```python
def plot_wave(wave, info):
    d = compute_delay(wave)
    tp, I = compute_db(wave)
    print('Duration:', d, '\n', 'Intensity:', I)
    plt.plot(wave, '|',label='wave')
    plt.axvline(x=d*48000*1.5, color='green', linestyle='--',label = 'Duration')
    plt.axhline(tp, color='red',linestyle='-', label = 'Intensity')
    plt.title(str(info)[1:-1])
```

```python
wave_audi, info_audi = wave(1, 'audi', 37, 25, 0.7)
wave_lab, info_lab = wave(1, 'lab', 37, 25, 0.7)
```

```python
plot_wave(wave_audi,info_audi)
IPython.display.Audio(wave_audi, rate=48000)
```

```python
plot_wave(wave_lab,info_lab)
IPython.display.Audio(wave_audi, rate=48000)
```
