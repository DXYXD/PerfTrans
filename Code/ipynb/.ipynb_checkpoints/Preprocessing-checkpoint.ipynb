### Compute DB

```python
# Computing decibel by time window with length t
# Average intensity of the ﬁrst phase of decay from the peak
def compute_db(frame, t, ref):
    decibel = []
    for i in frame:
        energy = np.array(i ** 2)
        onset = np.where(energy == max(energy))[0][0] # the index of the maximum energy
        tp = np.sqrt(np.mean(energy[onset:(onset + t)])) # # root mean square amplitude of the ﬁrst 10ms after the peak
        decibel.append(librosa.amplitude_to_db([tp], ref)[0])
        
    return decibel
```

### Compute Delay

```python
def compute_delay(frame, threshold, sr):
    delay = []
    length = sr//100
    for i in range(4*17*88):
        j = 10
        tp = threshold+1
        while(tp > threshold):
            j+=1
            tp = np.sqrt(np.var((frame[i][j*length:j*length+length])))
        delay.append(float(j)/100)
        
    return delay
```

```python
def loads(file_path, notetype, pitch, vel_level,vel, dur_level,dur):
    ''' sr = 48000 '''
    name = 'dictwave_'+notetype+'_pitch-'+str(pitch)+'_vel-'+str(vel[vel_level])+'_dur-'+str(dur[dur_level])+'_idx-'+str(dur_level+4*vel_level+4*len(vel)*(pitch-21))
    load_dict = np.load(file_path+'\\'+name+'.npy')
    print('type:',notetype, '  ', 'pitch:', pitch, '  ', 'velocity:', vel[vel_level], '  ', 
          'duration:', dur[dur_level], '  ', 'index:', dur_level+4*vel_level+4*len(vel)*(pitch-21))
    info = {'type': notetype, 'sr': sr, 'pitch':pitch, 'velocity':vel[vel_level], 
            'duration':dur[dur_level], 'index':dur_level+4*vel_level+4*len(vel)*(pitch-21)}
    data = load_dict
    
    return data, info
```
