```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
```

```python
csv_file = 'D:\\Academic_work\\00PerfTransfer\\File\\csv\\anal_new.csv'
df = pd.read_csv(csv_file)
so2_decibel = (np.array(df['db(lab)']))
so1_decibel = (np.array(df['db(audi)']))
so2_delay = np.array(df['duration(lab)'])
so1_delay = np.array(df['duration(audi)'])
so1_decibel.shape = (88,17,4)
so2_decibel.shape = (88,17,4)
so1_delay.shape = (88,17,4)
so2_delay.shape = (88,17,4)
```

```python
x = [0.1, 0.3,0.7,1]
y1 = so1_delay[36,7,:]
y2 = so2_delay[36,7,:]
ft1 = interp1d(x, y1, kind='linear')
ft2 = interp1d(x, y2, kind='linear')
```

```python
dd = 0.136458
s0 = ft1(dd) 
t1 = np.linspace(0.1, 1, 50)
s1 = ft2(t1)
diff = np.abs(s1-s0)
tstar = np.where(diff==min(diff))[-1]
t1[tstar][-1]
```

```python
import pretty_midi
# Create a PrettyMIDI object
cello_c_chord = pretty_midi.PrettyMIDI()
# Create an Instrument instance for a cello instrument
cello_program = pretty_midi.instrument_name_to_program('Cello')
cello = pretty_midi.Instrument(program=cello_program)
# Create a Note instance, starting at 0s and ending at .5s
note = pretty_midi.Note(
    velocity=60, pitch=57, start=27.419792, end=27.556250)
# Add the cello instrument to the PrettyMIDI object
cello.notes.append(note)
cello_c_chord.instruments.append(cello)
# Write out the MIDI data
cello_c_chord.write('cello-C-chord.mid')
```
